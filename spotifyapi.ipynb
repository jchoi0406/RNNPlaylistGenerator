```python
def authenticate(api_key, api_secret):
    auth_string = f"{api_key}:{api_secret}"
    encoded_auth_string = base64.b64encode(auth_string.encode()).decode()
    headers = {
        'Authorization': f'Basic {encoded_auth_string}'
    }
    data = {
        'grant_type': 'client_credentials'
    }
    
    response = requests.post('https://accounts.spotify.com/api/token', headers=headers, data=data)
    return response.json().get("access_token")
```

```python
def get_spotify_tracks(df, token, na_rows=None):
    total_rows = len(df)
    df_to_iterate = df
    if na_rows is not None:
        df_to_iterate = na_rows
        total_rows = len(na_rows)
        
    for index, row in tqdm(df_to_iterate.iterrows(), total=total_rows):
        query = f"{row['track_name']} {row['artist']}"
        url = f"https://api.spotify.com/v1/search?q={query}&type=track&limit=1"
        headers = {
            "Authorization": f"Bearer {token}"
        }
        
        try:
            response = requests.get(url, headers=headers)
            
            tracks = response.json().get('tracks', {}).get('items', [])
            if tracks:
                track_id = tracks[0].get('id')
                df.at[index, 'track_ids'] = track_id
            else:
                df.at[index, 'track_ids'] = None

        except Exception as e:
            print(f"Error fetching track ID for {query}: {e}")
            df.at[index, 'track_ids'] = None 
    
    return df

    

token = authenticate(api_key, api_secret)
new_df = get_spotify_tracks(df, token, na_rows)
print(new_df)
```

```python
audio_features = ["acousticness", "danceability", "duration_ms", "energy", "instrumentalness", "key", "liveness", "loudness", "speechiness", "tempo", "time_signature", "valence"]
```

```python
def fetch_features(ids_chunk, token):
    url = f"https://api.spotify.com/v1/audio-features?ids={','.join(ids_chunk)}"
    headers = {
        "Authorization": f"Bearer {token}"
    }
    response = requests.get(url, headers=headers)
    return response.json()
```

```python
def get_audio_features(df, token, na_rows=None):
    df_to_iterate = df
    if na_rows is not None:
        df_to_iterate = na_rows
    track_ids = df_to_iterate['track_ids'].tolist()
    chunks = [track_ids[i:i + 100] for i in range(0, len(track_ids), 100)]
    results = {}
    
    for chunk in tqdm(chunks):
        data = fetch_features(chunk, token)
        for item in data.get('audio_features', []):
            if item is not None:  # Check if item is not None
                track_id = item.get('id')
                if track_id:
                    results[track_id] = {feature: item.get(feature) for feature in audio_features}
                
    for index, row in tqdm(df_to_iterate.iterrows(), total=len(df_to_iterate)):
        track_id = row['track_ids']
        if track_id in results:
            df.loc[index, audio_features] = [results[track_id].get(feature) for feature in audio_features]
        else:
            df.loc[index, audio_features] = [None] * len(audio_features)  

    return df
        
    
```
